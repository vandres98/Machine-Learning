## Delta rule - weights update

```python
# Initial weights and learning rate
w = np.array([0.1,0.1,0.1])
mu = 0.1

print(f"Initial weights: {w}")
print(f"Initial mu: {mu}")

# AND Perceptron:
x1 = np.array([-1,-1,1,1])
x2 = np.array([-1,1,-1,1])
b = (1,1,1,1)
y = np.array([-1,-1,-1,1])

print("\nData:")
print(f"x1 = {x1}")
print(f"x2 = {x2}")
print(f"b = {b}")
print(f"y = {y}\n")

for epoche in range(2):
    print("=======================================")
    print(f"Epoche {epoche+1}:")
    counter = 1
    for i in [0,1,2,3]:
        print(f"Step {counter}:")
        d = np.array([x1[i],x2[i],b[i]])
        print(f"point {i}: {d}")
        t = y[i]
        print(f"target t: {t}")

        # outputi = w1x1 + w2x2 + w0*b
        sigma = np.dot(d, w)
        print(f"sigma = w1x1 + w2x2 + w0*b = {sigma}")
        output_i = 1 if sigma>0 else -1
        print(f"Heaviside output o = {output_i}")
        delta_w = [None,None,None]
        for i in range(len(d)):
            delta_w[i] = (mu * (t - output_i) * d[i])
        print(f"Delta_wi = mu * (t- output_i) * x_d[0]")
        print(f"Delta_w0 = {mu} * ({t} - {output_i}) * {d[0]} = {delta_w[0]}")
        print(f"Delta_w1 = {mu} * ({t} - {output_i}) * {d[1]} = {delta_w[1]}")
        print(f"Delta_w2 = {mu} * ({t} - {output_i}) * {d[2]} = {delta_w[2]}")

        #Update w:
        print(f"w = w + delta_w = {w} + {delta_w} = {w + delta_w}")
        w = w + delta_w
        
        counter += 1
        print("=======================================")
```
